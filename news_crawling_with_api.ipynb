한 리퀘스트 당 받아올 수 있는 양이 제한되어있어서 사이트 별로 나누어 진행한다.
- 한 리퀘스트 당 최대 100개 결과
- page를 100개 받아올 수 있도록 5개로 조정 
- 그런데 한 페이지만 출력된다... (20개)

```python
myapi = '85812a9149a343e28a1d8c36b8a846dd'

# url 만드는 함수
def makeurl(domain, q, myapi=myapi):
    url = "https://newsapi.org/v2/everything?domain={}&q={}&from=2021-02-06&sortBy=popularity&language=en&page=5&apiKey={}".format(domain, q, myapi)
    return url
```

```python
# request 후 받아온 내용에서 기사 제목만 반환
def request_and_make(site, query, myapi=myapi):
    response = requests.get(makeurl(site, query, myapi))
    news = response.json()
    return [i['title'] for i in news['articles']]
```

```python
site_list = ['forbes.com', 'nasdaq.com', 'wsj.com', 'reuters.com']
data = []

for i in site_list:
    data.extend(request_and_make(i, 'apple'))

print('headline 개수:', len(data))
print(data[:20])
```

```python
df = pd.DataFrame(data,columns=['news_headline'])
```
